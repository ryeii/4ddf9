```python
# read results
import pandas as pd
import matplotlib.pyplot as plt

PPO_result = pd.read_csv("PPO_results.csv")
NN_result = pd.read_csv("NN_results.csv")

# Plot the mean rewards
plt.plot(PPO_result["Timesteps_trained"], PPO_result["rewards_mean"])
plt.xlabel("Timesteps trained")
plt.ylabel("Mean rewards")
plt.title("Mean rewards over 10000 timesteps vs timesteps trained")

# draw a red dashed line at y = 0
plt.axhline(y=0, color='r', linestyle='--')

# draw a gray dashed line at y = -0.015
plt.axhline(y=-0.015, color='gray', linestyle='--')

# x axis limit
plt.xlim(0, 50000)

rewards_25 = PPO_result["rewards_25"]
rewards_75 = PPO_result["rewards_75"]
# fill the area between the upper and lower 2 std
plt.fill_between(PPO_result["Timesteps_trained"], rewards_25, rewards_75, color='b', alpha=.1)


plt.plot(NN_result["Timesteps_trained"], NN_result["rewards_mean"])
rewards_25 = NN_result["rewards_25"]
rewards_75 = NN_result["rewards_75"]
# fill the area between the upper and lower 2 std
plt.fill_between(NN_result["Timesteps_trained"], rewards_25, rewards_75, color='r', alpha=.1)


# add legends: red dashed line is the highest possible reward, gray dashed line is the Oracle+MPPI reward
plt.legend(["PPO mean rewards", "Highest possible mean reward", "Oracle+MPPI mean reward"])

plt.savefig("PPO_mean_rewards.png")
plt.show()
```
